## 학습 방법
ML 에는 모델과 그에 따른 Loss function 이 존재한다. Loss function 을 가지고 있는 데이터를 통해 구한 후, 이를 최소화 시키는 parameter 를 구한다. 이러한 과정을 최적화(Optimization) 이라 한다.

이 때 모델과 Loss function 은 복잡하기 때문에 일반적으로 Gradient Descent 와 같이 수치 해석적으로 최적화 시킨다.

### Gradient Descent
Loss function 이 Convex (아래로 볼록) 할 경우 이 방법으로 최적화 할 수 있다. 식은 다음과 같다.

$$ \sigma \to \sigma - \alpha{\delta \over {\delta w}}L(w) $$

이 때 $\alpha$ 를 learning rate 라 하며 이 값이 학습 속도를 결정 한다. 만약 값이 크면 학습 속도는 빠르겠지만 부정확 하고, 값이 작다면 반대이다.

이것 말고도 Gradient Descent 의 한계를 해결하기 위해서 확률적인 개념을 도입한 Stocastic Gradient Descent 와 같은 방법도 존재한다.

## ML 종류
### 레이블 유무에 따른 종류
- **지도 학습 (Supervised Learning)**

데이터의 레이블이 주어진 상태에서의 학습
- **비지도 학습 (Unsupervised Learning)**

데이터의 레이블이 주어지지 않은 상태에서의 학습

- **강화 학습**

#### 예시

영화 데이터로 영화의 관람객을 예측하기 위한 학습은 관람객이라는 레이블이 주어저 있으므로 지도 학습.

특정 데이터가 어떠한 형태로 군집을 형성하는지 알아보기 위한 학습은 비지도 학습.

## 평가

만든 모델이 얼마나 적합한지 평가를 하는 방법은 여러가지인데, 일반적으로는 데이터 셋을 학습을 위한 셋과 테스트를 위한 셋으로 나누어 평가를 진행한다.

학습을 위한 데이터를 이용하여 Gradient Descent 나 다른 여러가지 방법을 통해 모델을 최적화 시키고, 만든 모델을 테스트 셋에 적용 시켜 테스트 셋에 있는 결과 값을 모델이 얼마나 예측을 잘 하는지 평가하는 것이다.

이 때 데이터를 두 부분으로 독립적인 시행을 통해 나눈다면, 학습 데이터와 테스트 데이터 간 상관관계가 높지 않을 것이므로, 평가 결과를 신뢰할 수 있다.
